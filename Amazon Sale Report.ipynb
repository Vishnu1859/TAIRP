#### Dataset Describtion:
    
This dataset provides an in-depth look at the profitability of e-commerce sales.
It contains data on a variety of sales channels, including Shiprocket and INCREFF, as well as financial information on related expenses and profits. The columns contain data such as SKU codes, design numbers, stock levels, product categories, sizes and colors. In addition to this we have included the MRPs across multiple stores like Ajio MRP , Amazon MRP , Amazon FBA MRP , Flipkart MRP , Limeroad MRP Myntra MRP and PaytmMRP along with other key parameters like amount paid by customer for the purchase , rate per piece for every individual transaction Also we have added transactional parameters like Date of sale months category fulfilledby B2b Status Qty Currency Gross amt . This is a must-have dataset for anyone trying to uncover the profitability of e-commerce sales in today's marketplace

#### Attributes

* 'Order ID': Unique identifier for each order.
* 'Date': Date of the order.
* 'Status': Status of the order (e.g., processing, shipped, delivered).
* 'Fulfilment': Fulfillment status of the order.
* 'Sales Channel': The channel through which the sale was made (e.g., online platform, store).
* 'ship-service-level': Service level for shipping.
* 'Style': Style information for the product.
* 'SKU': Stock Keeping Unit, a unique identifier for each distinct product or item.
* 'Category': Product category.
* 'Size': Size of the product.
* 'ASIN': Amazon Standard Identification Number, a unique identifier for products on Amazon.
* 'Courier Status': Status of the courier service used for shipping.
* 'Qty': Quantity of items ordered.
* 'currency': Currency used for the transaction.
* 'Amount': Amount paid for the order.
* 'ship-city': City to which the order is shipped.
* 'ship-state': State or region to which the order is shipped.
* 'ship-postal-code': Postal code of the shipping address.
* 'ship-country': Country to which the order is shipped.
* 'promotion-ids': IDs related to promotions applied to the order.
* 'B2B': Indicates whether the order is a business-to-business transaction.
* 'fulfilled-by': Information about who fulfills the order (e.g., third-party, in-house).
* 'Unnamed: 22': An unnamed or unspecified column in the dataset. This column may require further investigation to understand its purpose.

### Data Cleaning

```python
amz_df.columns
```

```python
# # Removing unwanted columns

amz_df.drop(columns=['Order ID', 'Unnamed: 22'],axis=1,inplace=True)
```

```python
amz_df['B2B'].replace(to_replace=[True,False],value=['business','customer'], inplace=True)
```

```python
# Checking for duplicated values

len(amz_df.duplicated())
```

```python
amz_df.shape
```

```python
amz_df[amz_df.duplicated()]
```

```python
# Creating Datetime and adding Month column

amz_df['Date'] = pd.to_datetime(amz_df['Date'])
```

```python
amz_df['Month'] = amz_df['Date'].dt.month
amz_df["Month"].unique()
```

```python
amz_df['Month'].value_counts()
```

```python
month_dict = { 3: 'march', 4: 'april',5: 'may',6: 'june'}
amz_df['Month'] = amz_df['Date'].dt.month.map(month_dict)
```

```python
month = ['march', 'april', 'may', 'june']

amz_df['Month'] = pd.Categorical(amz_df['Month'], categories=month, ordered=True)
```

```python
amz_df.head()
```

```python
# Removing duplicated rows based on all columns

new_df = amz_df.drop_duplicates()
```

```python
new_df.duplicated().sum()
```

```python
# Check for Mising Values

new_df.isnull().sum()
```

```python
# Impute missing values with the median for numerical columns

new_df_num = new_df.select_dtypes(include=['number'])
```

```python
new_df_num.fillna(new_df_num.median(),inplace=True)
```

```python
new_df_num.isnull().sum()
```

```python
# Impute missing values with the most frequent category for categorical columns

new_df_cat = new_df.select_dtypes(include=['object'])
```

```python
for col in new_df_cat.columns:
    new_df_cat[col].fillna(new_df_cat[col].mode()[0],inplace=True)
```

```python
new_df_cat.isnull().sum()
```

```python
df = pd.concat([new_df_cat, new_df_num],axis = 1)
```

```python
df["customer_type"] = amz_df["B2B"]
```

```python
df["Month"] = amz_df["Month"]
```

```python
df.drop(columns=["B2B"], inplace=True)
```

### Bi-Variate Analysis

```python
# Shipping Status of Orders based on Category

cat_st = pd.crosstab(df['Category'],df['Status'])
```

```python
plt.figure(figsize=(20,20))
cat_st.plot(kind='bar', stacked=True, colormap='tab10')
plt.title('Status of Orders based on Category')
plt.xlabel('Category')
plt.ylabel('count')
plt.show()
```

* we can infer that most of the customers cancelled set and kurtha. this can be because of quality or dislike of the material
* there is a huge demand for set and kurta is high. the products are waiting for pickup
* most of the customers in the dataset are purchasing set and kurtha
* there are very few values in the shipping mode
* only few of the customers have been cancelled saree and bottom
* there are no customers who cancelled dupatta

```python
# Top 10 products based on Order Quantity

top_10P = df.groupby('Category')[['Qty']].max().sort_values(by='Qty',ascending = False)
plt.figure(figsize=[15,10])
sns.barplot(x=top_10P['Qty'],y=top_10P.index, palette='mako')
plt.title('Top 10 products based on Order Quantity')
plt.show()
```

* most of the customers purchased blouse and kurta
* few of the customers purchased dupatta

```python
# Top 10 expensive products

top_10_pa = df.groupby('Category')[['Amount']].max().sort_values(by='Amount',ascending = False).head(10)
plt.figure(figsize=[15,10])
sns.lineplot(x=top_10_pa['Amount'], y=top_10_pa.index, palette='Set2')
plt.title('Top 10 expensive products')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
plt.show()
```

* Set and Western dress are the most expensive products while dupatta is the most cheapest product.
